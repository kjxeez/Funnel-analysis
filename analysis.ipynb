```python
df = pd.read_csv('data.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df['month'] = df['date'].dt.month
```

```python
df['week'] = df['date'].dt.week
```

```python
df1=df.groupby('month').agg({'clients': 'sum', 'accounts':'sum', 'applications':'sum', 'games':'sum'})
df1
```

```python
df2=df.groupby('week').agg({'clients': 'sum', 'accounts':'sum', 'applications':'sum', 'games':'sum'})
df2
```

```python
from plotly import graph_objects as go
```

```python
# создаем воронку для первой таблицы
funnel_data = pd.DataFrame({'labels': ['Visits', 'Registrations', 'Applications', 'Games'],
'values': [df1['clients'].sum(), df1['accounts'].sum(),
df1['applications'].sum(), df1['games'].sum()]})

# добавляем столбцы с долей каждого этапа и его накопленной долей
funnel_data['cumulative_values'] = funnel_data['values'].cumsum()
funnel_data['stage_perc'] = round(funnel_data['values'] / funnel_data['values'][0] * 100, 2)
funnel_data['cumulative_perc'] = round(funnel_data['cumulative_values'] / funnel_data['values'].sum() * 100, 2)
```

```python
# создаем воронку для второй таблицы
funnel_data2 = pd.DataFrame({'labels': ['Visits', 'Registrations', 'Applications', 'Purchases'],
'values': [df2['clients'].sum(), df2['accounts'].sum(),
df2['applications'].sum(), df2['games'].sum()]})

# добавляем столбцы с долей каждого этапа и его накопленной долей
funnel_data2['cumulative_values'] = funnel_data2['values'].cumsum()
funnel_data2['stage_perc'] = round(funnel_data2['values'] / funnel_data2['values'][0] * 100, 2)
funnel_data2['cumulative_perc'] = round(funnel_data2['cumulative_values'] / funnel_data2['values'].sum() * 100, 2)
```
