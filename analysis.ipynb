```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df['month'] = df['date'].dt.month
df
```

```python
df.groupby('month').agg({'clients': 'sum', 'accounts':'sum', 'applications':'sum', 'games':'sum'})
```
